```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns

import plotly.express as px
```

```python
iris = pd.read_csv("sample_data/iris.csv")
iris.head()
```

```python
iris["variety"].value_counts()
```

```python
# Scatter Plotting using seaborn

sns.set(style="white", color_codes=True)
iris = sns.load_dataset('iris')

sns.FacetGrid(iris, hue ="species",
              height = 6).map(plt.scatter,
                              'sepal_length',
                              'petal_length').add_legend()

# sns.FacetGrid(iris, hue ="species",
#               height = 6).map(plt.scatter,
#                               'species',
#                               'petal_length').add_legend()

plt.show()
```

```python
# Scatterplotting using matplotlib 

iris.plot(kind="scatter", x="sepal.length", y="petal.length", color="#007bff", s=60)
# iris.plot(kind="scatter", x="variety", y="petal.length", color="#007bff", s=60)

plt.show()
```

```python
# Box Plotting using seaborn


sns.boxplot(x="species", y="petal_length", data=iris )
plt.show()
```

```python
# Histogram

iris['sepal.length'].plot(kind='hist')
# iris['sepal.length'].hist()
plt.show()
```

```python
# Density Chart

iris['sepal.length'].plot(kind='density')
plt.show()
```

```python
# Andrews Curves

x = pd.plotting.andrews_curves(iris, 'variety')
  
x.plot()
plt.show()
```

```python
# Parallel Plotting


df = px.data.iris()
fig = px.parallel_coordinates(df, color="species_id", labels={"species_id": "Species",
                "sepal_width": "Sepal Width", "sepal_length": "Sepal Length",
                "petal_width": "Petal Width", "petal_length": "Petal Length", },
                             color_continuous_scale=px.colors.diverging.Tealrose,
                             color_continuous_midpoint=2)
fig.show()
```

```python
# Bubble Chart with plotly.express

x = iris["sepal.length"].to_numpy()
y = iris["petal.length"].to_numpy()
z = iris["sepal.length"].to_numpy()


# x = np.random.rand(40)
# y = np.random.rand(40)
# z = np.random.rand(40)
colors = np.random.rand(len(z)) 

plt.scatter(x, y,c=colors)
# plt.scatter(x, y, s=z*1000)

plt.show()
```

```python
# Bubble Chart with plotly.graph_objects

import plotly.graph_objects as go

fig = go.Figure(data=[go.Scatter(
    x=iris["sepal.length"].to_numpy(),
     y=iris["petal.length"].to_numpy(),
    mode='markers')
])

fig.show()
```
